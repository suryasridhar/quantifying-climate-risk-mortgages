```python
# ── STEP 1: Mount Google Drive ──────────────────────────────────────────────
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

# ── STEP 2: Load both CSVs from Drive ───────────────────────────────────────
# Update these paths to match your actual file locations
df1 = pd.read_csv('/content/drive/MyDrive/FEMA_Datasets/FEMA_dataset.csv')
df2 = pd.read_csv('/content/drive/MyDrive/FEMA_Datasets/IndividualAssistanceMultipleLossFloodProperties.csv')

# ── STEP 3: Inspect columns ──────────────────────────────────────────────────
print("=== df1 columns ===")
print(df1.columns.tolist())
print("\n=== df2 columns ===")
print(df2.columns.tolist())

# Optional: see first few rows of each
df1.head()
```

```python
print(df1.columns.tolist())
```

```python
# First verify what columns actually exist
print("Actual df1 columns:", df1.columns.tolist())
print("df1 shape:", df1.shape)
```

```python
df1_cols_keep = [
    'disasterNumber',
    'state',
    'zipCode',
    'totalDamage',
    'totalApprovedIhpAmount'
]

# Check which of your desired columns are actually missing before selecting
missing = [col for col in df1_cols_keep if col not in df1.columns]
print("Missing columns:", missing)

# Only proceed if no missing columns
df1_clean = df1[df1_cols_keep].copy()

# Normalize ZIP to 5-digit string
df1_clean['zipCode'] = df1_clean['zipCode'].astype(str).str.zfill(5)

print("df1_clean shape:", df1_clean.shape)
df1_clean.head()
```

```python
# Rename zip and state columns for consistency
df2 = df2.rename(columns={'damagedZipCode': 'zipCode', 'stateAbbreviation': 'state'})

# Columns to keep (floodInsurance dropped, residenceType brought back)
df2_cols_keep = [
    'disasterNumber', 'declarationDate',
    'state', 'zipCode', 'county', 'city',
    'highRiskPropertyType', 'residenceType',
    'destroyed', 'foundationType', 'waterLevel',
    'floodDamage', 'numberOfLosses'
]
df2_clean = df2[df2_cols_keep].copy()

# Normalize ZIP
df2_clean['zipCode'] = df2_clean['zipCode'].astype(str).str.zfill(5)

# ── Filter residenceType to only A, C, H, TH ─────────────────────────────────
valid_residence_types = ['A', 'C', 'H', 'TH']

# Strip whitespace first to avoid mismatch on accidental spaces
df2_clean['residenceType'] = df2_clean['residenceType'].astype(str).str.strip().str.upper()

before = len(df2_clean)
df2_clean = df2_clean[df2_clean['residenceType'].isin(valid_residence_types)]
after = len(df2_clean)

print(f"Rows removed due to residenceType filter: {before - after}")
print(f"df2_clean shape after filter: {df2_clean.shape}")
print("Remaining residenceType values:\n", df2_clean['residenceType'].value_counts())
df2_clean.head()
```

```python
# ── Aggregate df1 to ZIP level ────────────────────────────────────────────────
df1_zip = df1_clean.groupby('zipCode').agg(
    total_damage         = ('totalDamage',            'sum'),
    total_ihp_approved   = ('totalApprovedIhpAmount', 'sum'),
    disaster_count_df1   = ('disasterNumber',         'nunique'),
    state                = ('state',                  'first')
).reset_index()

print("df1_zip shape:", df1_zip.shape)
df1_zip.head()
```

```python
# ── Aggregate df2 to ZIP level ────────────────────────────────────────────────
# Encode destroyed column if stored as Yes/No strings
df2_clean['destroyed'] = (
    df2_clean['destroyed']
    .astype(str).str.strip().str.upper()
    .map({'YES': 1, 'NO': 0, 'TRUE': 1, 'FALSE': 0})
    .fillna(0).astype(int)
)

df2_zip = df2_clean.groupby('zipCode').agg(
    total_flood_damage   = ('floodDamage',    'sum'),
    avg_flood_damage     = ('floodDamage',    'mean'),
    total_losses         = ('numberOfLosses', 'sum'),
    avg_losses_per_event = ('numberOfLosses', 'mean'),
    destroyed_count      = ('destroyed',      'sum'),
    avg_water_level      = ('waterLevel',     'mean'),
    disaster_count_df2   = ('disasterNumber', 'nunique'),
    total_events         = ('disasterNumber', 'count')
).reset_index()

print("df2_zip shape:", df2_zip.shape)
df2_zip.head()
```

```python
# Install pgeocode (lightweight, fully offline, no API key needed)
!pip install pgeocode -q

import pgeocode
import re

# ── Layer 1: Format validation (must be exactly 5 numeric digits) ─────────────
def is_valid_format(zip_code):
    return bool(re.fullmatch(r'\d{5}', str(zip_code)))

# ── Layer 2: Check against real US ZIP database ───────────────────────────────
nomi = pgeocode.Nominatim('us')

def is_valid_us_zip(zip_code):
    result = nomi.query_postal_code(str(zip_code))
    return pd.notna(result['latitude'])   # invalid ZIPs return NaN coords

# ── Validate unique ZIPs only (much faster than row-by-row) ──────────────────
unique_zips_df1 = df1_clean['zipCode'].unique()
unique_zips_df2 = df2_clean['zipCode'].unique()

all_unique_zips = pd.Series(
    list(set(unique_zips_df1) | set(unique_zips_df2))
).astype(str)

print(f"Total unique ZIPs to validate: {len(all_unique_zips)}")

# Step 1: format filter
format_valid = all_unique_zips[all_unique_zips.apply(is_valid_format)]
format_invalid = all_unique_zips[~all_unique_zips.apply(is_valid_format)]
print(f"Format invalid ZIPs: {len(format_invalid)}")
print("Examples:", format_invalid.values[:10])

# Step 2: database check on format-valid ZIPs
db_valid_mask = format_valid.apply(is_valid_us_zip)
db_valid_zips  = format_valid[db_valid_mask]
db_invalid_zips = format_valid[~db_valid_mask]
print(f"\nDatabase invalid ZIPs: {len(db_invalid_zips)}")
print("Examples:", db_invalid_zips.values[:10])

# Final set of clean ZIPs
valid_zips = set(db_valid_zips.values)
print(f"\nValid ZIPs remaining: {len(valid_zips)}")
```

```python
# ── Apply filter to both cleaned dataframes ───────────────────────────────────
before_df1 = len(df1_clean)
before_df2 = len(df2_clean)

df1_clean = df1_clean[df1_clean['zipCode'].isin(valid_zips)]
df2_clean = df2_clean[df2_clean['zipCode'].isin(valid_zips)]

print(f"df1: {before_df1} → {len(df1_clean)} rows ({before_df1 - len(df1_clean)} dropped)")
print(f"df2: {before_df2} → {len(df2_clean)} rows ({before_df2 - len(df2_clean)} dropped)")
```

```python
# ── Now proceed with ZIP-level aggregation and merge (same as before) ─────────
df1_zip = df1_clean.groupby('zipCode').agg(
    total_damage         = ('totalDamage',            'sum'),
    total_ihp_approved   = ('totalApprovedIhpAmount', 'sum'),
    disaster_count_df1   = ('disasterNumber',         'nunique'),
    state                = ('state',                  'first')
).reset_index()

df2_clean['destroyed'] = (
    df2_clean['destroyed']
    .astype(str).str.strip().str.upper()
    .map({'YES': 1, 'NO': 0, 'TRUE': 1, 'FALSE': 0})
    .fillna(0).astype(int)
)

df2_zip = df2_clean.groupby('zipCode').agg(
    total_flood_damage   = ('floodDamage',    'sum'),
    avg_flood_damage     = ('floodDamage',    'mean'),
    total_losses         = ('numberOfLosses', 'sum'),
    avg_losses_per_event = ('numberOfLosses', 'mean'),
    destroyed_count      = ('destroyed',      'sum'),
    avg_water_level      = ('waterLevel',     'mean'),
    disaster_count_df2   = ('disasterNumber', 'nunique'),
    total_events         = ('disasterNumber', 'count')
).reset_index()

fema_merged = pd.merge(df1_zip, df2_zip, on='zipCode', how='outer')
fema_merged.fillna(0, inplace=True)

print("Final merged shape:", fema_merged.shape)
fema_merged.head()
```
